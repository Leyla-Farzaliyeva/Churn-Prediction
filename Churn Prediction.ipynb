```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
data = pd.read_csv("C:\\Users\\hp\\Desktop\\BOOTCAMP\\Python\\Week 8\\Telco_Customer_Churn.csv")
data
```

```python
data.info()
```

```python
data['Churn'].replace(to_replace='Yes', value=1, inplace=True)
data['Churn'].replace(to_replace='No',  value=0, inplace=True)
```

```python
X = data.drop(columns = ['Churn'])
X
```

```python
Y = data['Churn']
Y
```

```python
#transforming categorical variables into dummy variables
X = pd.get_dummies(X)
X
```

```python
#splitting data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
#Scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#SVM w/ Linear kernel
from sklearn.svm import SVC
classifier_l = SVC(kernel = "linear", random_state=0)
classifier_l.fit(X_train, Y_train)
```

```python
y_pred = classifier_l.predict(X_test)
y_pred
```

```python
from sklearn.metrics import  accuracy_score, classification_report, confusion_matrix
cm = metrics.confusion_matrix(Y_test, y_pred) 
print(cm)
accuracy_l = metrics.accuracy_score(Y_test, y_pred) 
print("Accuracy score:",accuracy_l)
precision = metrics.precision_score(Y_test, y_pred) 
print("Precision score:",precision)
recall = metrics.recall_score(Y_test, y_pred) 
print("Recall score:",recall)
```

```python
#SVM w/ RBF kernel
classifier_rbf = SVC(kernel = "rbf", random_state=0)
classifier_rbf.fit(X_train, Y_train)
```

```python
y_pred = classifier_rbf.predict(X_test)
y_pred
```

```python
cm = metrics.confusion_matrix(Y_test, y_pred) 
print(cm)
accuracy_rbf = metrics.accuracy_score(Y_test, y_pred) 
print("Accuracy score:",accuracy_rbf)
precision = metrics.precision_score(Y_test, y_pred) 
print("Precision score:",precision)
recall = metrics.recall_score(Y_test, y_pred) 
print("Recall score:",recall)
```

```python
#Decision Tree classifier
from sklearn.tree import DecisionTreeClassifier
classifier_dt = DecisionTreeClassifier(criterion = "entropy", random_state=0)
classifier_dt.fit(X_train, Y_train)
```

```python
y_pred = classifier_dt.predict(X_test)
y_pred
```

```python
cm = metrics.confusion_matrix(Y_test, y_pred) 
print(cm)
accuracy_dt = metrics.accuracy_score(Y_test, y_pred) 
print("Accuracy score:",accuracy_dt)
precision = metrics.precision_score(Y_test, y_pred) 
print("Precision score:",precision)
recall = metrics.recall_score(Y_test, y_pred) 
print("Recall score:",recall)
```

```python
#RandomForest Classifier

from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(n_estimators=100, criterion="gini", random_state=0)
classifier_rf.fit(X_train, Y_train)
```

```python
y_pred = classifier_rf.predict(X_test)
y_pred
```

```python
cm = metrics.confusion_matrix(Y_test, y_pred) 
print(cm)
accuracy_rf = metrics.accuracy_score(Y_test, y_pred) 
print("Accuracy score:",accuracy_rf)
precision = metrics.precision_score(Y_test, y_pred) 
print("Precision score:",precision)
recall = metrics.recall_score(Y_test, y_pred) 
print("Recall score:",recall)
```

```python
model_list=["SVM(Linear Kernel)", "SVM(RBF Kernel)", "Decision Tree", "RandomForest"]
accuracy_list = [accuracy_l, accuracy_rbf, accuracy_dt, accuracy_rf]
scores_df = pd.DataFrame()

scores_df["Accuracy"]=accuracy_list
scores_df.index=model_list

scores_df.sort_values(by=["Accuracy"], ascending=False)
```

```python
#10-fold Cross Validation on SVM w/linear kernel

from sklearn.model_selection import cross_val_score 
scores = cross_val_score(classifier_l, X_train, Y_train, cv=10)
mean_l = scores.mean()
print('CV scores:', scores)
print('Mean Accuracy:',mean_l)
```

```python
#10-fold Cross Validation on SVM w/ RBF kernel
from sklearn.model_selection import cross_val_score 
scores = cross_val_score(classifier_rbf, X_train, Y_train, cv=10)
mean_rbf = scores.mean()
print('CV scores:', scores)
print('Mean Accuracy:',mean_rbf)
```

```python
#10-fold Cross Validation on Decision Tree classifier
from sklearn.model_selection import cross_val_score 
scores = cross_val_score(classifier_dt, X_train, Y_train, cv=10)
mean_dt = scores.mean()
print('CV scores:', scores)
print('Mean Accuracy:',mean_dt)
```

```python
#10-fold Cross Validation on Random Forest classifier
from sklearn.model_selection import cross_val_score 
scores = cross_val_score(classifier_rf, X_train, Y_train, cv=10)
mean_rf = scores.mean()
print('CV scores:', scores)
print('Mean Accuracy:',mean_rf)
```

```python
model_list=["SVM(Linear Kernel)", "SVM(RBF Kernel)", "Decision Tree", "RandomForest"]
accuracy_list = [mean_l, mean_rbf, mean_dt, mean_rf]
scores_df = pd.DataFrame()

scores_df["Mean Accuracy of 10-fold CV"]=accuracy_list
scores_df.index=model_list

scores_df.sort_values(by=["Mean Accuracy of 10-fold CV"], ascending=False)
```
